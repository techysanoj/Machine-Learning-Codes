```python
sample_text = "Data structures and algorithm is one of the most important topics for your placement and interview preparation. Here is a complete roadmap to cover all the data structure algorithm topics and make you comfortable with the most important algorithms in data structure for your interviews. You can practice the company-specific data structure algorithm questions and give your career a push."
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize
```

```python
# to get all the sentences from the given text

sentences = sent_tokenize(sample_text)
sentences
```

```python
# to get all the words fromt the given words

word = word_tokenize(sample_text)
word
```

```python
# there are many stop words which are use less to train the model
# to get all the stopwords we will use 

from nltk.corpus import stopwords
stop =  stopwords.words('english')
stop
```

```python
# as all stops words are in lowercase so, while doing word tokenize we need to convert it into lowercase
# but it is not true that doing this we may lost the word info.

word = word_tokenize(sample_text.lower())
word
```

```python
# to get all the clean words we will use for loop over the word file

clean_word = [w for w in word if not w in stop]

clean_word
```

```python
# as we can see here we have so many punctuation present so to remove this we will add punctutation in stop words

import string
punctuation = list(string.punctuation)
punctuation
```

```python
stop+=punctuation
```

```python
clean_words2 = [w for w in word if w not in stop]
clean_words2
```
