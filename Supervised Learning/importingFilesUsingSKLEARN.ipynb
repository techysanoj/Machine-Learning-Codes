```python
from sklearn import datasets

boston = datasets.load_boston()
X = boston.data
Y = boston.target
```

```python
import pandas as pd

df = pd.DataFrame(X)
df.columns = boston.feature_names
df["age_age"] = df.AGE**2
df["rad_tax"] = df.RAD*df.TAX
df["nox_rm"] = df.NOX*df.RM

X2 = df.values
df.describe()
```

```python
from sklearn import model_selection
X_train, X_test, Y_train, Y_test = model_selection.train_test_split(X,Y, random_state = 0)
new_xtrain, new_xtest, new_ytrain, new_ytest = model_selection.train_test_split(X2, Y, random_state = 0)
```

```python
#importing algorithm
from sklearn.linear_model import LinearRegression
algo = LinearRegression()
algo2 = LinearRegression()
algo.fit(X_train,Y_train)

algo2.fit(new_xtrain, new_ytrain)

train_score1 = algo.score(X_train, Y_train)
test_score1 = algo.score(X_test, Y_test)
print("Train Score 1:", train_score1)
print("Test Score 1:", test_score1)
train_score2 = algo2.score(new_xtrain, new_ytrain)
test_score2 = algo2.score(new_xtest, new_ytest)
print("Train Score 2:", train_score2)
print("Test Score 2:", test_score2)
```

```python
Y_pred = algo.predict(X_test)
Y2_pred = algo2.predict(new_xtest)
```

```python
# to compare the Y_pred and Y_test
import matplotlib.pyplot as plt
plt.scatter(Y_test,Y_pred)
plt.scatter(new_ytest, Y2_pred)
plt.axis([0,40,0,40])
plt.show()
```
