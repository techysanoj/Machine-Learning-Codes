```python
import numpy as np

data = np.loadtxt("data.csv", delimiter = ",")

# make an one d array of all ones and append it into data
d_ones = np.ones(len(data)).reshape(-1,1)

# make an one d array of ml 
ml = np.array(data[0:, 1]).reshape(-1,1)

new_data = np.append(data, d_ones, axis = 1)
new_data = np.append(new_data, ml, axis = 1)
new_data
```

```python
def cost(data, m):
    total_cost = 0
    y = data[0:, data.shape[1]-1]
    for i in len(data):
        total_cost += (1/len(data))*(((new_data[i]*m).sum() -  y[i])**2)
    return total_cost
        
```

```python
def step_gradient(data, learning_rate, m):
    m_slope = np.zeros(data.shape[1], dtype = int)
    y = data[0:, data.shape[1]-1]
    M = len(data)
    for i in range(M):
        y = data[0:, data.shape[1]-1]
        for j in range(len(m_slope)):
            m_slope[j] += (-2/M)*((y[i] - (new_data[i]*m).sum())*new_data[i][j]
        new_m = m - learning_rate*m_slope
    return new_m
```

```python
def gd(data, learning_rate, num_iteration):
    m = np.zeros(data.shape[1], dtype = int)
    for i in range(num_iteration):
        m = step_gradient(data, learning_rate, m)
    return m
```

```python
def run():
    learning_rate = 0.0001
    num_iteration = 10
    m = gd(data, learning_rate, num_iteration)
    print(m)
```

```python
run()
```
